```python
import numpy as np
import pandas as pd

dataset = pd.read_csv("adult.csv", header = None)
dataset
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.columns = ["age", "workclass", "wgt", "edu", "edu_num", "marital_status",
                   "occ", "relation", "race", "sex", "gain", "loss", "hpw", "country", "income"]
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy = "most_frequent")
dataset1 = si.fit_transform(dataset)
dataset1 = pd.DataFrame(dataset1)
dataset1.isna().sum()
```

```python
dataset1.head()
```

```python
#label encoding the data
cols = [1,3,5,6,7,8,9,13,14]
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in cols:
  dataset1[i] = le.fit_transform(dataset1[i])
dataset1.head()
```

```python
target = dataset1[14]
input = dataset1.drop(columns = [14])
print(target.shape)
print(input.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input,target,test_size = 0.2)
print(x_train.shape)
print(x_test.shape)
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
#implementing perceptron
from sklearn.linear_model import Perceptron
pn = Perceptron()
pn.fit(x_train, y_train)
train_results = pn.predict(x_train)
test_results = pn.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Training Accuracy: ",accuracy_score(y_train, train_results))
print("Testing Accuracy: ",accuracy_score(y_test, test_results))
```

Using confusion Matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train,train_results)
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score
precision_score(y_train, train_results)
```

```python
recall_score(y_train, train_results)
```

```python
f1_score(y_train, train_results)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr_train_pred = lr.predict(x_train)
lr_test_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Training Accuracy: ",accuracy_score(y_train, lr_train_pred))
print("Testing Accuracy: ",accuracy_score(y_test, lr_test_pred))
```

SVM

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train,y_train)
train_results=svc.predict(x_train)
test_result=svc.predict(x_test)
y_predict=svc.predict(x_test)
from sklearn.metrics import accuracy_score
print("Accuracy: %.2f" %accuracy_score(y_test, y_predict))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(criterion='entropy')
dt.fit(x_train,y_train)
train_results=dt.predict(x_train)
test_result=dt.predict(x_test)
y_predict=svc.predict(x_test)
from sklearn.metrics import accuracy_score
print("Accuracy: %.2f" %accuracy_score(y_test, y_predict))
```
